```python
import networkx as nx
import matplotlib.pyplot as plt

# 예시 그래프 생성
G = nx.Graph()

# 노드 추가
G.add_node("Node1")
G.add_node("Node2")
G.add_node("Node3")
G.add_node("Node4")
G.add_node("Node5")

# 엣지 추가
G.add_edge("Node1", "Node2")
G.add_edge("Node2", "Node3")
G.add_edge("Node3", "Node1")
G.add_edge("Node4", "Node2")
G.add_edge("Node5", "Node2")
G.add_edge("Node1", "Node4")

# 그래프 시각화
pos = nx.spring_layout(G)  # 레이아웃 설정
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=1000, node_color='skyblue', font_size=10, edge_color='gray', width=2)

plt.title('그래프 시각화 예제')
plt.show()
```

```python
category = pd.read_excel("C:\\Users\\user\\Desktop\\데이터분석 부트캠프\\6. SQL\\SQL 프로젝트\\카테고리별 관계.xlsx", sheet_name = 2, header = 0)
category
```

```python
import networkx as nx

G = nx.from_pandas_edgelist(category, source='category1_eng', target='category2_eng',
                            edge_attr='cnt')
```

```python
import matplotlib.pyplot as plt

node_size_factor = 100
k_factor = 50

cnt = np.array([w for *_, w in G.edges.data('cnt')])

# Figure 크기 조절
plt.figure(figsize=(18, 12))

pos = nx.spring_layout(G, seed=7)  # positions for all nodes - seed for reproducibi

# nodes
nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=700)

# edges
nx.draw_networkx_edges(G, pos, edge_color='gray', width=cnt/7)  # using a 10x scale factor here

# labels
nx.draw_networkx_labels(G, pos, font_size=10, font_family="sans-serif")

ax = plt.gca()
ax.margins(0.08)
plt.axis("off")
plt.tight_layout()
```
